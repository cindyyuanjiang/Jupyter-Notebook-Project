```python
#-------Importing Libraries-------------
import pandas as pd
import seaborn as sns
from numpy import nan
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
f=pd.read_csv('bitcoin_dataset.csv')
```

```python
f.head(1024)
```

```python
f.corr()
```

```python
sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
sns.jointplot(data=f,x=f["btc_market_price"],y=f["btc_market_cap"])
```

```python
#---------Replacing all 0.000 values with nan------------
f.replace(0,np.nan,inplace=True)
```

```python
#----------Data Cleaning-----------------
df=f[["btc_market_price","btc_market_cap","btc_n_transactions","btc_miners_revenue","btc_cost_per_transaction","btc_difficulty","btc_hash_rate","btc_cost_per_transaction_percent"]]
```

```python
df.head()
```

```python
#-----------Ignoring the savedcopywarning error---------------
pd.options.mode.chained_assignment = None
```

```python
#--------------replacing nan values with their respective column mean-------------
df.fillna(df.mean()["btc_market_price":"btc_market_price"],inplace=True)
df.fillna(df.mean()["btc_market_cap":"btc_market_cap"],inplace=True)
df.fillna(df.mean()["btc_n_transactions":"btc_n_transactions"],inplace=True)
df.fillna(df.mean()["btc_miners_revenue":"btc_miners_revenue"],inplace=True)
df.fillna(df.mean()["btc_cost_per_transaction":"btc_cost_per_transaction"],inplace=True)
df.fillna(df.mean()["btc_difficulty":"btc_difficulty"],inplace=True)
df.fillna(df.mean()["btc_hash_rate":"btc_hash_rate"],inplace=True)
df.fillna(df.mean()["btc_cost_per_transaction_percent":"btc_cost_per_transaction_percent"],inplace=True)
```

```python
df.head()
```

```python
X=df[["btc_market_cap","btc_n_transactions","btc_miners_revenue","btc_cost_per_transaction","btc_difficulty","btc_hash_rate","btc_cost_per_transaction_percent"]]
X.head()
```

```python
Y=df[["btc_market_price"]]
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
lm.coef_
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(x = Y_test, y = predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt
rmse=mean_squared_error(Y_test,predictions,squared=False)
```
